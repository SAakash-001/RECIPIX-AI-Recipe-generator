```python
# Importing Libraries

import requests
from urllib.parse import unquote
```

```python
# Using Spenacular API to generate Food Recipe from list
API_KEY = '6d8987a709a741c9bcd23f6ac569b499'
```

```python
def get_recipe_information(recipe_id):
    url = f'https://api.spoonacular.com/recipes/{recipe_id}/information'
    params = {
        'apiKey': API_KEY,
    }

    # Send a GET request to the Spoonacular API with the query parameters
    response = requests.get(url, params=params)
    # If the API call is successful
    if response.status_code == 200:
        # Parse the API response as JSON data
        return response.json()
    # If the API call is not successful
    return None
```

```python
def search_recipes_by_ingredients(ingredients):
    url = 'https://api.spoonacular.com/recipes/findByIngredients'
    params = {
        'apiKey': API_KEY,
        'ingredients': ','.join(ingredients),
        'number': 10,
        'ranking': 1,  # Adjust the ranking parameter as needed
    }

    # Send a GET request to the Spoonacular API with the query parameters
    response = requests.get(url, params=params)
    # If the API call is successful
    if response.status_code == 200:
        # Parse the API response as JSON data
        data = response.json()
        # Return the list of recipe results
        return data
    # If the API call is not successful
    return []
```

```python
# Example of using the search_recipes_by_ingredients function
ingredient_list = ['chicken', 'onion', 'garlic']
recipes = search_recipes_by_ingredients(ingredient_list)
```

```python
# Print detailed information for each recipe
for recipe in recipes:
    recipe_id = recipe['id']
    recipe_info = get_recipe_information(recipe_id)

    if recipe_info:
        print(f"Recipe ID: {recipe_info['id']}")
        print(f"Recipe Title: {recipe_info['title']}")
        print(f"Used Ingredients: {', '.join(ingredient['name'] for ingredient in recipe['usedIngredients'])}")
        print(f"Instructions: {recipe_info['instructions']}")
        print("")
```
